```
import pandas as pd

#read in the data
campaign = pd.read_csv('https://raw.githubusercontent.com/byui-cse/cse450-course/master/data/bank.csv')
#bank = pd.read_csv('https://raw.githubusercontent.com/byui-cse/cse450-course/master/data/bank.csv')
bank_holdout = pd.read_csv('https://raw.githubusercontent.com/byui-cse/cse450-course/master/data/bank_holdout_test.csv')
bank_hold_mini = pd.read_csv('https://raw.githubusercontent.com/byui-cse/cse450-course/master/data/bank_holdout_test_mini.csv')
```

```
#decision tree
from sklearn import tree
from sklearn.ensemble import RandomForestClassifier
#from sklearn.preprocessing import OneHotEncorder
from sklearn.model_selection import train_test_split

#outcome target (y-variable) = customers most likely to subscribe to a term deposit
campaign.head(5)

campaign['campaign'] = campaign['campaign'].astype('category')

#encode features and target variables
features = ['emp.var.rate',
            'euribor3m',
            'cons.price.idx',
            'nr.employed',
            'contact'
            #'campaign', 'poutcome', 'age',
            #'education', # interested to know if they should contact students
            #'marital', #single people on Saturdays? #better to visualiz with graph?
            #'day_of_week' #focus on Saturday
            ]
X = pd.get_dummies(campaign[features])
y = campaign['y']

#split to train and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20)

#build decision tree
clf = RandomForestClassifier()

#train
clf.fit(X_train, y_train)

#test
clf.score(X_test, y_test)
```

```
#PREDICT ON THE MINI DATASET
#assign the same features to the mini dataset
mini_X = pd.get_dummies(bank_hold_mini[features])

#align columns with the campaign dataset
mini_X = bank_hold_mini.reindex(columns = X.columns, fill_value=0)

#make a predictions column, filled in with the predictions
mini_X['predictions (y)'] = clf.predict(mini_X)

mini_X
```

```
#PERFORMANCE OF MODEL
#accuracy, precision, recall, F1 score
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
y_pred = clf.predict(X_test)

# calculations
precision = precision_score(y_test, y_pred, pos_label='yes', average='binary')
recall = recall_score(y_test, y_pred, pos_label='yes', average='binary')
f1 = f1_score(y_test, y_pred, pos_label='yes', average='binary')

print(f"Precision Score: {precision}")
print(f"Recall: {recall}")
print(f"F1 Score: {f1}")


```
